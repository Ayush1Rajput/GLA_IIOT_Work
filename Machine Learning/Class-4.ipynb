```python
from sklearn.datasets import make_regression
```

```python
import numpy as np
```

```python
X,y=make_regression(n_samples=4, n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# let apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Let apply Gradient Descent assuming slope is constant m=78.35
# and Let's assume the starting value for intercept b=100
y_pred=((78.35*X)+100).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
m=78.35
b=100
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# Let take leasrning rate=0.1
lr=0.1
step_size=loss_slope*lr
step_size
```

```python
# Calculating the new intercept
b=b-step_size
b
```

```python
y_pred1=((78.35*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='yellow',label='b={}'.format(b))
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
lr=0.1
step_size=loss_slope*lr
print(step_size)
b=b-step_size
b
```

```python
y_pred2=((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='yellow',label='b=40.92')
plt.plot(X,y_pred2,color='aqua',label='b={}'.format(b))
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```
