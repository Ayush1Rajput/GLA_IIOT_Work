```python
# Import necessary libraries


import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Load dataset 
data = pd.read_csv('iot_device_data.csv')
```

```python
# Quick look at data

# print(data.head(30))

data.sample(30)
```

```python
#Define the features and the target 
# Assuming the target is 'status' and it's categorical te.g., "normal" or "faulty)
X =  data[['temperature', 'humidity', 'pressure', 'vibration']] #Use relevant Iof features 
y = data['status']
```

```python
#Convert categorical target to numerical (if necessary) 
y= pd. factorize(y)[0]
```

```python
#Split the dataset into training and testing sets 
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

#Initialize the Random Forest Classifier 
model = RandomForestClassifier(n_estimators=20, random_state=42)
model
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```
