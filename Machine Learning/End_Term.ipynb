```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = iris.data
y_true = iris.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

K-Means Clustering 

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
y_kmeans = kmeans.labels_

silhouette_avg = silhouette_score(X_scaled, y_kmeans)
print("For n_clusters =", 3,
      "The average silhouette_score is :", silhouette_avg)

plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, cmap='viridis')
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Agglomerative Clustering

```python
agg = AgglomerativeClustering(n_clusters=3)
agg.fit(X_scaled)
y_agg = agg.labels_

silhouette_avg = silhouette_score(X_scaled, y_agg)
print("For n_clusters =", 3,
      "The average silhouette_score is :", silhouette_avg)


plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_agg, cmap='viridis')
plt.title("Agglomerative Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

DBSCAN Algorithm

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(X_scaled)
y_dbscan = dbscan.labels_

silhouette_avg = silhouette_score(X_scaled, y_dbscan)
print("For the above set of parameters, the average silhouette_score is :", silhouette_avg)

# Visualize the clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_dbscan, cmap='viridis')
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
